```python
precio= 20000
def mostrar_producto(nombre, color, tipo, precio):
    print("El Producto es", nombre)
    print("El color es", color)
    print("Es de tipo", tipo)
    print("Cuesta ", precio)
print("Cual es el nombre del producto?")
nombre= input("Cual es el nombre del producto?")
print("Cual es el color del producto?")
color= input("Cual es el color del producto?")
print("Cual es el tipo de producto?")
tipo= input("Cual es el tipo?")

mostrar_producto(nombre, color, tipo, precio)
```
