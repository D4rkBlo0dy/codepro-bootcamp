Escribir un programa que almacene una contrase침a en una variable, y pregunte al usuario por la contrase침a hasta que introduzca la contrase침a correcta.

Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

```python
# Almacenamos la contrasena

contrasena = "12345"
while True:
    contrasena_ingresada = input("ingrese la contrasena")
    if contrasena == contrasena_ingresada:
        print("la contrasena es correcta")
        break
    else:
        print("la contrasena es incorrecta")
```

### Crear una lista de 5 elementos. Luego, seguir los siguientes pasos.

Agregar un elemento al final de la lista.
Modificar el segundo elemento.
Eliminar el tercer elemento.

```python
#creamos una lista

frutas = ["naranja", "banana", "manzana", "melon", "kiwi"]
print(frutas)

#Agregar un elemento al final

frutas.append("durazno")
print(frutas)

#modificar el segundo elemento

frutas[1] = "ciruela"
print(frutas)

#eliminar el tercer elemento

frutas.pop(2)
print(frutas)
```

Escribir un programa que itere en un rango de 10 veces. Por cada iteracion multiplicar el iterador por dos e imprimir el resultado.

### Challenge 游뱤 - 7m
Recorrer los elementos de una lista de marcas e imprimir cada marca solamente si tiene 6 letras o m치s.

```python
lista_de_marcas=["Nike", "Adidas", "Converse", "Vans", "Reebok"]
 
for marcas in lista_de_marcas:
    if len(marcas) >= 6:
        print(marcas)
```
