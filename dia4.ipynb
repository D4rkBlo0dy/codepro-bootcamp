## Challenge 2 內 - 20 min

Crear una tarjeta personal. El usuario (mediante `input`) puede elegir:

1. `Crear` datos nuevos
2. `Modificar` datos
3. `Eliminar` datos

```python
#Crear una tarjeta personal
tarjeta_personal= {}
while True:
    opcion= int(input("Elige la opcion que desea usar 1- Crear Datos, 2- Modificar datos, 3-Eliminar datos, 4- Salir"))
    if opcion == 1:
        clave= input("ingrese la clave a guardarse")
        valor= input("ingrese el valor de la clave")
        tarjeta_personal[clave]= valor
        print(tarjeta_personal)
    elif opcion == 2:
        print("Estos son las claves y valores que puedes modificar")
        print(tarjeta_personal.items())
        clave= input("ingrese la clave a cambiarse")
        valor= input("ingrese el valor de la clave a cambiarse")
        tarjeta_personal[clave]= valor
        print(tarjeta_personal)
    elif opcion == 3:
        print("Estos son los valores que puedes eliminar")
        print(tarjeta_personal.keys())
        clave= input("ingrese la clave a eliminar")
        del tarjeta_personal[clave]
        print(tarjeta_personal)
    elif opcion == 4:
        break
```
